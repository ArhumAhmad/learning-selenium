```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
driver = webdriver.Firefox()

driver.get('https://www.worldometers.info/world-population/population-by-country/')
xp = "/html/body/div[3]/div[2]/div/div/div[2]/table/tbody/tr/td[{0}]"

countries     = driver.find_elements(by=By.XPATH, value = xp.format(2))
time.sleep(1)
populations   = driver.find_elements(by=By.XPATH, value = xp.format(3))
time.sleep(1)
yearly_change = driver.find_elements(by=By.XPATH, value = xp.format(4))
time.sleep(1)
world_share   = driver.find_elements(by=By.XPATH, value = xp.format(12))
time.sleep(1)
```

```python
population_result = []
for i in range(len(yearly_change)):
    temp_data = {}
    temp_data['Country']       = countries[i].text
    temp_data['Populations']   = populations[i].text
    temp_data['Yearly Change'] = yearly_change[i].text
    temp_data['World Share']   = world_share[i].text
    population_result.append(temp_data)

driver.close()
```

```python
df = pd.DataFrame(population_result)
df
```

```python
df.to_excel('world_pops.xlsx', sheet_name='sheet1')
```
